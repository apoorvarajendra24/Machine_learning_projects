```python
# Multicolinearity check
def correlation(dataset, threshold):
    corr_matrix = dataset.corr()
    high_corr_pairs = []
    to_drop = set()
    # Loop over lower triangle of the matrix
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            corr_value = abs(corr_matrix.iloc[i, j])
            if corr_value > threshold:
                col1 = corr_matrix.columns[i]
                col2 = corr_matrix.columns[j]
                high_corr_pairs.append((col1, col2))
                # Add one of the two columns to drop list
                to_drop.add(col1)

    return high_corr_pairs, to_drop
# Threshold -- set by domain expert
pairs, drop_col = correlation(df, 0.8)
print("Highly correlated pairs...")
for p in pairs:
    print(p)
print("Suggested columns to drop...")
print(drop_col)
```

# Feature Engineering

```python
df['Rate_to_income'] = (df['MonthlyRate'] / df['MonthlyIncome'])
#This is the most common method.
#It shows what portion of income is spent on the rate.
#Values >1 mean the rate is more than the income (possibly unsustainable).
#Values <1 are more financially sound.
```

```python
#Income relative to hours — helps normalize.
#Monthly income and hourly rate both talk about salary, but separately they don’t tell you efficiency. 
#Combining them shows if someone is underpaid or overpaid compared to their hourly work — a strong indicator for employee dissatisfaction or attrition.
df['Income_per_hour'] = (df['MonthlyIncome'] / (df['HourlyRate'] * 30))
```

```python
#Stability indicator — years spent per company.
#Tells you the average years spent per company. If someone changes jobs too often (low value), 
#it signals instability, which is very predictive of attrition (quitting) risk.

df['Working_years_per_company'] = np.where(
    df['NumCompaniesWorked'] == 0,
    df['TotalWorkingYears'],  # assume all experience is from 1 company
    df['TotalWorkingYears'] / df['NumCompaniesWorked'])
```

```python
df['Working_years_per_company'].isin([np.inf, -np.inf]).any()
```

```python
#Frequent promotions? High value to company.
#Measures how often a person gets promoted. 
#Employees who wait too long without promotion might become unhappy. Adding 1 prevents division by zero. Very predictive for attrition.
df['Promotion_frequency'] = (df['TotalWorkingYears'] / (df['YearsSinceLastPromotion'] + 1))
```

```python
#How distance affects satisfaction.
#Employees who live far and are less satisfied may quit. 
#This feature captures the stress of commuting vs. happiness with the job. Again, +1 prevents division by zero.
df['Distance_JobSatisfaction']  = (df['DistanceFromHome'] / (df['JobSatisfaction'] + 1))
```

```python
#Overall satisfaction with work and personal life.
#WorkLifeBalance measures work pressure, and RelationshipSatisfaction measures workplace social happiness.
#Combining them gives an overall “emotional satisfaction” score — very useful in predicting attrition.
df['WorkLife_Satisfaction'] = (df['WorkLifeBalance'] + df['RelationshipSatisfaction']) / 2
```

```python
#Loyalty or working stability.
#Measures loyalty/stability with current manager compared to time in role. 
#If someone has a bad manager or keeps changing managers, it affects satisfaction and risk of quitting.
df['YearsInRole_ManagerRatio'] = (df['YearsWithCurrManager'] / (df['YearsInCurrentRole'] + 1))
```

```python
#JobLevel shows seniority; Education shows qualification. Combining them checks if someone's job matches their education level. 
#A mismatch (e.g., highly educated but low job level) can predict dissatisfaction or attrition.
df['JobLevel_Education'] = (df['JobLevel'] * df['Education'])
```

```python
#Measures how long someone stayed at a company relative to age — indicates loyalty. Older people with short tenures may be job hoppers.
df['JobStability_Score'] = (df['YearsAtCompany'] / (df['Age'] + 1))
```

```python
#Combines involvement + satisfaction. If both are low, employee is disengaged and likely to quit.
df['Engagement_Score'] = ((df['JobInvolvement'] + df['EnvironmentSatisfaction']) / 2)
```

```python
#If someone performs well but gets a low raise, this may cause dissatisfaction. Captures reward mismatch.
df['PerformanceToReward'] = (df['PerformanceRating'] / (df['PercentSalaryHike'] + 1))
```

```python
#Indicates how much the company is investing in the person. More training + benefits = more retention.
df['SkillInvestment_Score'] = (df['TrainingTimesLastYear'] * df['StockOptionLevel'])
```

```python
#binary feature showing likely overworked and dissatisfied employee. Useful for logistic regression.
df['Stress_score'] = np.where((df['OverTime'] == 1) & (df['EnvironmentSatisfaction'] < 3),1,0)
```

```python
# Delete the columns now which are combined with others to create a new feature.
d = ['EmployeeCount','EmployeeNumber','Over18','Over18_encoded','StandardHours','MonthlyIncome','HourlyRate','TotalWorkingYears','NumCompaniesWorked','YearsSinceLastPromotion','DistanceFromHome','JobSatisfaction','WorkLifeBalance','RelationshipSatisfaction','YearsWithCurrManager','YearsInCurrentRole']
```

```python
df.drop(d,axis=1, inplace=True)
```

```python
# Deleting the features used in New Feature Engineering
df.drop(columns=['YearsAtCompany','Age','JobInvolvement','EnvironmentSatisfaction','PerformanceRating','PercentSalaryHike','TrainingTimesLastYear','StockOptionLevel','OverTime'],axis=1, inplace=True)
```

# Handling Outliers and using smote

```python
# Removing Outliers
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for each numerical column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_cleaned = df[~((df < lower_bound) | (df > upper_bound)).any(axis=1)]

# Resulting cleaned DataFrame
df_cleaned
```

```python
df = df_cleaned.copy()
```

```python
from imblearn.over_sampling import SMOTE

X = df.drop('Attrition_encoded', axis=1)
y = df['Attrition_encoded']
```

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# Check for infinity values
print(np.isinf(df).sum())  # Count infs in each column
print(np.isnan(df).sum())  # Count NaNs in each column
```

```python
# Recreate a DataFrame
df_resampled = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), 
                          pd.DataFrame(y_resampled, columns=['Attrition_encoded'])], axis=1)
```

```python
df = df_resampled.copy()
```
